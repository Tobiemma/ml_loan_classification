# Remplir les valeurs catÃ©gorielles manquantes avec le modus

```python
#VÃ©rifier les valeurs manquantes
cat_data.isnull().sum().sort_values(ascending=False)
```

```python
cat_data.head()
```

```python
#Remplir les valeurs manquantes de Gender par le modus

mode_gender = cat_data['Gender'].mode().values[0]
cat_data['Gender'] = cat_data['Gender'].fillna(mode_gender, inplace= False)
cat_data['Gender'].isnull().sum()
```

```python
#Remplir les valeurs manquantes de Self_Employed par le modus

mode_self_employed = cat_data['Self_Employed'].mode().values[0]
cat_data['Self_Employed'] = cat_data['Self_Employed'].fillna(mode_self_employed, inplace = False)
cat_data['Self_Employed'].isnull().sum()
```

```python
#Remplir les valeurs manquantes de Married par le modus

mode_married = cat_data['Married'].mode().values[0]
cat_data['Married'] = cat_data['Married'].fillna(mode_married, inplace = False)
cat_data['Married'].isnull().sum()
```

```python
cat_data.isnull().sum().sort_values(ascending=False)
```

# Remplir les valeurs numeriques manquantes avec la mediane

```python
num_data.isnull().sum().sort_values(ascending=False)
```

```python
num_data = df.select_dtypes(np.number)
num_data
```

```python
med= num_data['Credit_History'].median()
med
```

```python
#Remplir les valeurs manquantes de Credit_History par la mediane
numerische_spalten = ['Credit_History', 'LoanAmount', 'Dependents', 'Loan_Amount_Term']

# Fehlende Werte in numerischen Spalten mit dem Median auffÃ¼llen
for spalte in numerische_spalten:
    med= num_data[spalte].median()
    num_data[spalte] = num_data[spalte].fillna(med, inplace=False)
```
