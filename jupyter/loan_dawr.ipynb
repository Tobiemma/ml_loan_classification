### Managing incorrectly recorded values

```python
# verify unique values
#df.Dependents.unique()
df['Dependents'].unique()
```

```python
# replace object '3+' values with int 3 values
df['Dependents'].replace('3+', 3, inplace=True)
```

```python
# verify unique values
df['Dependents'].unique()
```

```python
#Convert object values to float64
df['Dependents']= df['Dependents'].astype(float)
```

```python
df.info()
```

```python
# categorie & num
cat_data = df.select_dtypes(object)
num_data = df.select_dtypes(np.number)
```

```python
cat_data.head()
```

```python
num_data.head()
```

### Fill in missing categorical values with modus

```python
#Check missing values
cat_data.isnull().sum().sort_values(ascending=False)
```

```python
cat_data.head()
```

```python
# Fill Gender's missing values with the modus

mode_gender = cat_data['Gender'].mode().values[0]
cat_data['Gender'] = cat_data['Gender'].fillna(mode_gender, inplace= False)
cat_data['Gender'].isnull().sum()
```

```python
# Fill missing Self_Employed values with modus

mode_self_employed = cat_data['Self_Employed'].mode().values[0]
cat_data['Self_Employed'] = cat_data['Self_Employed'].fillna(mode_self_employed, inplace = False)
cat_data['Self_Employed'].isnull().sum()
```

```python
# Fill in missing Married values with modus

mode_married = cat_data['Married'].mode().values[0]
cat_data['Married'] = cat_data['Married'].fillna(mode_married, inplace = False)
cat_data['Married'].isnull().sum()
```

```python
cat_data.isnull().sum().sort_values(ascending=False)
```

### Fill in missing numerical values with the median

```python
num_data.isnull().sum().sort_values(ascending=False)
```

```python
num_data = df.select_dtypes(np.number)
num_data
```

```python

numerische_spalten = ['Credit_History', 'LoanAmount', 'Dependents', 'Loan_Amount_Term']

# Fill missing values in numerical columns with the median
for spalte in numerische_spalten:
    med= num_data[spalte].median()
    num_data[spalte] = num_data[spalte].fillna(med, inplace=False)
```

```python
print(num_data)
```

```python
num_data['Dependents']= num_data['Dependents'].astype(int)
```

```python
print(num_data)
```

```python
num_data['Credit_History']= num_data['Credit_History'].astype(int)
```

```python
print(num_data)
```

```python
num_data.isnull().sum().sort_values(ascending=False)
```

```python
num_data
```

```python
cat_data
```

### Encoding

```python
cat_data_hot = pd.get_dummies(cat_data, columns = ['Gender', 'Married','Education', 'Self_Employed', 'Property_Area', 'Loan_Status'], dtype=int)
cat_data_hot                             
```

```python
df_clean= pd.concat([num_data, cat_data_hot], axis=1)
```

```python
df_clean
```

```python
df_clean.to_csv('../data_clean/data_train_clean.csv', index=False)
```
